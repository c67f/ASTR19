```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#randNums = np.random.hypergeometric(80, 560, 140, 1000)
randNums = np.random.pareto(2, 1000)
#print(randNums)

#width = 1
histmin = np.floor(min(randNums))
#print(histmin)
histmax = np.ceil(max(randNums))#+width
#print(histmax)

bins = np.linspace(histmin, histmax, 100)#np.arange(histmin, histmax)
#print(bins.size)
plt.hist(randNums, bins=bins)
plt.ylabel = "numbers per bin"
plt.xlabel = "bins"
plt.savefig("CodeJournal10Fig.pdf", bbox_inches="tight", facecolor="white")
```
