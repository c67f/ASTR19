```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000

randNums = np.random.randn(n)

#width = 1
histmin = np.floor(min(randNums))
#print(histmin)
histmax = np.ceil(max(randNums))#+width
#print(histmax)

bins = np.linspace(histmin, histmax, 100)#np.arange(histmin, histmax)
#print(bins)
plt.hist(randNums, bins=bins)
plt.ylabel = "numbers per bin"
plt.xlabel = "bins"
plt.savefig("CodeJournal9Fig.png", bbox_inches="tight", facecolor="white")
```
